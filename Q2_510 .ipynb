```python
df_new = pd.DataFrame(arr, columns =['branch-load-misses',
 'branch-misses',
 'iTLB-load-misses',
 'dTLB-load-misses',
 'dTLB-store-misses',
 'L1-icache-load-misses',
 'L1-dcache-load-misses',
 'l1d_pend_miss.pending_cycles',
 'l2_rqsts.code_rd_miss',
 'l2_rqsts.all_demand_miss',
 'l2_rqsts.demand_data_rd_miss',
 'l1d_pend_miss.pending_cycles',
 'l2_rqsts.code_rd_miss',
 'l2_rqsts.rfo_miss',
 'cache-misses',
 'LLC-load-misses',
 'LLC-store-misses',
 'l1d_pend_miss.l2_stall',
 'l2_rqsts.swpf_miss',
 'instructions',
 'cycles',
 'L1-dcache-loads',
 'L1-dcache-load-misses1',
 'LLC-loads',
 'LLC-load-misses1'])
df_new = df_new.replace('<not counted>','0')  
del df_new['L1-dcache-load-misses1']
del df_new['LLC-load-misses1']
del df_new['L1-dcache-loads']
del df_new['LLC-loads']
del df_new['instructions']
del df_new['cycles']

df_new
```

```python
li2=[]
for i in df['G']:
    li2.append(i)
li3=[]
for i in range(19,len(li2),25):
    li3.append(li2[i])
df_new['CPI']=li3
df_new
```

```python
# print(df_new['CPI'].astype(float).mean())
for i in list(df_new):
    df_new[i]=df_new[i].replace('0',df_new[i].astype(float).mean())

df_new
```
